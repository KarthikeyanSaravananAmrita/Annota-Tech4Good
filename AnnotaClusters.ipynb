```python
sentence_prediction = []
with jsonlines.open('/Users/karthikeyansaravanan/Downloads/drive-download-20230705T191437Z-001/7_3_2023 Sentence Predictions.jsonl', 'r') as reader:
    for line in reader:
        sentence_prediction.append(line)
```

```python
# Research Question ID
research_question_id = "Lxx8ibk76YOj51ypuyTu"
```

```python
# Filter relevant sentences based on the research question ID
relevant_sentences = []
for entry in sentence_prediction:
    if entry['u__researchQuestionId'] == research_question_id:
        transcript_id = entry['u__transcriptId']
        sentence_chunks = entry['u_sentenceChunks']
        relevant_pairs = []
        for chunk in sentence_chunks:
            is_relevant = chunk.get('isRelevant', False)
            if is_relevant:
                start_line = int(chunk['start'])
                end_line = int(chunk['end'])
                for line in range(start_line, end_line + 1):
                    relevant_pairs.append((transcript_id, line))
        relevant_sentences.extend(relevant_pairs)
        
# Print the relevant sentence pairs
for pair in relevant_sentences:
    transcript_id, line = pair
    print("Transcript ID:", transcript_id)
    print("Line:", line)
```

```python
# Print the relevant sentence pairs
for sentence in relevant_sentences:
    transcript_id, line = sentence
print("# of Relevant Sentences:", len(relevant_sentences))
```

```python
# Assign an index to each pair of sentences
sentence_index = {}
index = 0
for i1 in range(len(relevant_sentences)):
    sent1 = relevant_sentences[i1]
    for i2 in range(i1 + 1, len(relevant_sentences)):
        sent2 = relevant_sentences[i2]
        sentence_index[(sent1, sent2)] = index
        index += 1
```

```python
# Get the unique author IDs from theme versions, excluding 'DEFAULT'
author_ids = set(theme_version['u__authorId'] for theme_version in theme_versions if theme_version['u__authorId'] != 'DEFAULT')
```

```python
# Create the annotator index
author_index = {author_id: i for i, author_id in enumerate(author_ids)}
```

```python
# Calculate the number of relevant sentences
n = len(relevant_sentences)
```

```python
# Calculate the number of pairs
I = len(sentence_index.keys()) ## CORRECTION: I is not equal to n x n - 1, as (l1, l2) is the same task as (l2, l1)
J = 2
K = len(author_index.values())
```

```python
# Initialize the N matrix
N = [[[0 for _ in range(K)] for _ in range(J)] for _ in range(I)]
```

```python
# Process theme versions and annotations
for theme_version in theme_versions:
    if theme_version['u__researchQuestionId'] == research_question_id:
        # Store all annotations for this themeVersion, as well as which theme each one is assigned to.
        tv_annotations = []
        annotation_to_theme = dict()

        for theme in theme_version["structure"]:
            # Store all annotations from this theme and its subthemes for this themeVersion:
            tv_annotations += theme["u__annotationIds"]
            for subtheme in theme["children"]:
                tv_annotations += subtheme["u__annotationIds"]

            # Assign all annotations from this theme and its subthemes to this theme:
            for ann in theme["u__annotationIds"]:
                annotation_to_theme[ann] = theme

            for subtheme in theme["children"]:
                for ann in subtheme["u__annotationIds"]:
                    annotation_to_theme[ann] = theme

        # For every annotation A in the themeVersion
        for annotation_id in tv_annotations:
            annotation = next((ann for ann in annotations if ann['u__id'] == annotation_id), None)
            if annotation:
                transcript_id = annotation['u__transcriptId']
                start_line = int(annotation['startLine'])
                end_line = int(annotation['endLine'])

                # For x in range(A.startline, A.endline)
                for x in range(start_line, end_line):
                    # For y in range(A.startline, A.endline)
                    for y in range(start_line, end_line):
                        if x == y:
                            continue

                        pair_x = (transcript_id, x)
                        pair_y = (transcript_id, y)

                        # Skip if x or y is not relevant
                        if pair_x not in relevant_sentences or pair_y not in relevant_sentences:
                            continue

                        i = sentence_index[pair_x]
                        j = sentence_index[pair_y]

                        for author_id in author_ids:
                            k = author_index.get(author_id, -1)
                            if k != -1:
                                N[i][1][k] = 1
                                N[i][0][k] = 0

        # For every pair of annotations (A, B) in the themeVersion
        for i in range(len(tv_annotations)):
            for j in range(i + 1, len(tv_annotations)):
                annotation_id1 = tv_annotations[i]
                annotation_id2 = tv_annotations[j]
                annotation1 = next((ann for ann in annotations if ann['u__id'] == annotation_id1), None)
                annotation2 = next((ann for ann in annotations if ann['u__id'] == annotation_id2), None)
                if annotation1 and annotation2:
                    transcript_id1 = annotation1['u__transcriptId']
                    transcript_id2 = annotation2['u__transcriptId']
                    start_line1 = int(annotation1['startLine'])
                    end_line1 = int(annotation1['endLine'])
                    start_line2 = int(annotation2['startLine'])
                    end_line2 = int(annotation2['endLine'])

                    # For x in range(A.startline to A.endline)
                    for x in range(start_line1, end_line1):
                        # For y in range(B.startLine to B.endLine)
                        for y in range(start_line2, end_line2):
                            pair_x = (transcript_id1, x)
                            pair_y = (transcript_id2, y)

                            # Skip if x or y is not relevant
                            if pair_x not in relevant_sentences or pair_y not in relevant_sentences:
                                continue

                            i = sentence_index[pair_x]
                            j = sentence_index[pair_y]

                            for author_id in author_ids:
                                k = author_index.get(author_id, -1)
                                if k != -1:
                                    N[i][1][k] = 1
                                    N[i][0][k] = 1
```
