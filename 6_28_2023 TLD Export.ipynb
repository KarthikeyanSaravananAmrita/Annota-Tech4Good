```python
import json

def load_jsonl(file):
    with open(file, encoding="utf-8") as f:
        data = [json.loads(line) for line in f]
    return data

# Load the TLDs from the JSONL file
tlds = load_jsonl("/Users/karthikeyansaravanan/Downloads/Theme Discovery EM_ Anonymized Data/6_28_2023 TLD Export.json")

# Print the keys in the first entry
if tlds:
    first_entry = tlds[0]
    keys = first_entry.keys()
    print("Field names:", list(keys))
else:
    print("No data found in the JSONL file.")
```
